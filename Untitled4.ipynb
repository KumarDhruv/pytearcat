```python
import pytearcat as pt
```

```python
# Define the coordinates to be used
t,r,theta,phi = pt.def_coords('t,r,theta,phi')
# Define any constant (more than one constant can be defined at the same time)
a = pt.new_con('a')
c = pt.new_con('c')
M = pt.new_con('M')
G0 = pt.new_con('G0')
g = pt.create_metric('ds2 = -(1-(2*G0*M*r)/(c**2*(r**2+a**2/c**2*cos(theta)**2)))*c**2*dt**2 - (4*a*G0*M*r*sin(theta)**2)/(c**3*(r**2+a**2/c**2*cos(theta)**2))*c*dt*dphi+(r**2+a**2/c**2*cos(theta)**2)/(r**2-2*G0*M*r/c**2+a**2/c**2)*dr**2+(r**2+a**2/c**2*cos(theta)**2)*dtheta**2+((r**2+a**2/c**2+2*a**2*M*r*sin(theta)**2/(c**4*(r**2+a**2/c**2*cos(theta)**2)))*sin(theta)**2)*dphi**2')
```

```python
Chr = pt.christoffel()
```
